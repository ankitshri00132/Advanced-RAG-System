```python
class HybridSearcher :
    DENSE_MODEL = 'BAAI/bge-small-en-v1.5'
    SPARSE_MODEL = 'Qdrant/bm42-all-minilm-l6-v2-attentions'

    def __init__(self,collection_name):
        self.collection_name = collection_name
        self.qdrant_client = QdrantClient(url='http://localhost:6333')

    # search function

    def search(self,text : str):
        search_result = self.qdrant_client.query_points(
            collection_name = self.collection_name,
            query = models.FusionQuery(fusion=models.Fusion.RRF),
            prefetch = [
                models.Prefetch(
                    query = models.Document(text = text,model = self.DENSE_MODEL),
                    using = dense_vector_name,
                    limit = 10
                ),
                models.Prefetch(
                    query = models.Document(text = text,model= self.SPARSE_MODEL),
                    using = sparse_vector_name,
                    limit = 10
                ),
            ],
            limit = 10
        ).points

        metadata = []
        for point in search_result:
            metadata.append({
                'score' : point.score,
                'metadata':point.payload
            })
        return metadata
```

```python
from typing import List, Dict, Any
from fastembed.rerank.cross_encoder import TextCrossEncoder

def get_reranked_documents(
    query: str, 
    retrieved_results: List[Dict[str, Any]], 
    reranker_model: str = "jinaai/jina-reranker-v1-tiny-en"
) -> List[Dict[str, Any]]:
    """
    Reranks initial retrieved search results and fetches them in sorted order.
    
    Args:
        query: The search query.
        retrieved_results: List of search results from hybrid search containing 'payload' or 'metadata'.
        reranker_model: The name of the cross-encoder model to use.
        
    Returns:
        A list of dictionaries containing rank, score, and the document contents.
    """
    if not retrieved_results:
        return []
        
    # 1. Initialize the reranker
    reranker = TextCrossEncoder(reranker_model)
    
    # 2. Extract the document text from the retrieved results
    documents = []
    for item in retrieved_results:
        payload = item.get("payload") or item.get("metadata", {})
        # Check both standard Qdrant 'document' key and general 'text' key
        doc_text = payload.get("document") or payload.get("text", "")
        documents.append(doc_text)
        
    # 3. Get the new similarity scores from the reranker
    rerank_scores = reranker.rerank(query, documents)
    
    # 4. Generate ranks with their original index: [(original_index, rerank_score), ...]
    ranked_pairs = list(enumerate(rerank_scores))
    ranked_pairs.sort(key=lambda x: x[1], reverse=True)
    
    # 5. Fetch the original documents in ranked order and structure the output
    final_ranked_docs = []
    for rank_idx, (original_index, score) in enumerate(ranked_pairs, start=1):
        original_item = retrieved_results[original_index]
        payload = original_item.get("payload") or original_item.get("metadata", {})
        
        final_ranked_docs.append({
            "rank": rank_idx,
            "rerank_score": float(score),
            "original_score": original_item.get("score"),
            "document": payload.get("document") or payload.get("text", ""),
            "metadata": {k: v for k, v in payload.items() if k not in ["document", "text"]}
        })
        
    return final_ranked_docs
```
